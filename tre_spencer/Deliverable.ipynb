```python
# tried to code chart that where we can select a individual stock
import panel as pn

# Create a list of stock names
stocks = ['AMD', 'HD', 'NVDA'] 
# Create a dropdown widget with the stock names as options
stock_dropdown = pn.widgets.Select(name='Stock', options=stocks)

# Create a function that updates the plot based on the selected stock
def update(event):
    # Get the selected stock
    stock = event.new
    # Update the data in the plot to show data for the selected stock
    # df_stock_data.hvplot(x='Price ',
    #                      y='Market Cap',
    #                      kind='line', size=20,
    #                      color='green, purple',
    #                      data=df_stock_data[df_stock_data['Stock'] == stock])

# Link the dropdown widget to the update function
stock_dropdown.param.watch(update, 'value')

# Display the dropdown widget and the plot
pn.Column(stock_dropdown, df_stock_data.head(20).hvplot(x='Symbol',
                                                y='Price/Earnings',
                                               kind='scatter', size=200,
                                                color='Blue')).servable()
```
